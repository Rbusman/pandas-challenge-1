## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df['subtotal'] = df['unit_price'] * df['qty']
list(df.columns)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.


df['shipping_price'] = df['unit_weight'].apply(lambda x: 7 if x > 50 else 10) * df['unit_weight']

df.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%

df['total_price'] = df['subtotal'] + df['shipping_price'] * 0.0925

list(df.columns)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

df['cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
list(df.columns)
```

```python
# Create a column for the profit of
# each line using line cost and line price

df['total_profit'] = df['unit_price'] - df['unit_cost']
list(df.columns)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
dataset = [
    {'order_id': 1, 'client_id': 1001, 'total_price': 150.50},
    {'order_id': 2, 'client_id': 1002, 'total_price': 75.25},
    {'order_id': 3, 'client_id': 1003, 'total_price': 100.30},
    {'order_id': 4, 'client_id': 1002, 'total_price': 87.50},
    {'order_id': 5, 'client_id': 1001, 'total_price': 200.00},
    {'order_id': 6, 'client_id': 1004, 'total_price': 423.75},
    {'order_id': 7, 'client_id': 1003, 'total_price': 150.50},
    {'order_id': 8, 'client_id': 1002, 'total_price': 75.25},
    {'order_id': 9, 'client_id': 1001, 'total_price': 1000.50},
    {'order_id': 10, 'client_id': 1005, 'total_price': 500.00}
]  


client_spending = {}
for order in dataset:
    client_id = order['client_id']
    total_price = order['total_price']
    
    if client_id in client_spending:
        client_spending[client_id] += total_price
    else:
        client_spending[client_id] = total_price

# Sort clients by spending in descending order
top_clients = sorted(client_spending.items(), key=lambda x: x[1], reverse=True)[:5]

# Display the spending for each of the top 5 clients
for client_id, spending in top_clients:
    print(f"Client ID {client_id} spent ${spending}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.


dataset = [
    {'order_id': 1, 'client_id': 1001, 'total_units': 10, 'shipping_price': 5.00, 'revenue': 100.00, 'profit': 50.00},
    {'order_id': 2, 'client_id': 1002, 'total_units': 5, 'shipping_price': 2.50, 'revenue': 50.00, 'profit': 25.00},
    {'order_id': 3, 'client_id': 1003, 'total_units': 8, 'shipping_price': 4.00, 'revenue': 80.00, 'profit': 40.00},
    {'order_id': 4, 'client_id': 1002, 'total_units': 3, 'shipping_price': 1.50, 'revenue': 30.00, 'profit': 15.00},
    {'order_id': 5, 'client_id': 1001, 'total_units': 12, 'shipping_price': 6.00, 'revenue': 120.00, 'profit': 60.00},
    {'order_id': 6, 'client_id': 1004, 'total_units': 7, 'shipping_price': 3.50, 'revenue': 70.00, 'profit': 35.00},
    {'order_id': 7, 'client_id': 1003, 'total_units': 9, 'shipping_price': 4.50, 'revenue': 90.00, 'profit': 45.00},
    {'order_id': 8, 'client_id': 1002, 'total_units': 4, 'shipping_price': 2.00, 'revenue': 40.00, 'profit': 20.00},
    {'order_id': 9, 'client_id': 1001, 'total_units': 15, 'shipping_price': 7.50, 'revenue': 150.00, 'profit': 75.00},
    {'order_id': 10, 'client_id': 1005, 'total_units': 6, 'shipping_price': 3.00, 'revenue': 60.00, 'profit': 30.00}
]

# Calculate totals for each client
client_totals = {}
for order in dataset:
    client_id = order['client_id']
    total_units = order['total_units']
    shipping_price = order['shipping_price']
    revenue = order['revenue']
    profit = order['profit']
    
    if client_id in client_totals:
        client_totals[client_id]['total_units'] += total_units
        client_totals[client_id]['shipping_price'] += shipping_price
        client_totals[client_id]['revenue'] += revenue
        client_totals[client_id]['profit'] += profit
    else:
        client_totals[client_id] = {
            'total_units': total_units,
            'shipping_price': shipping_price,
            'revenue': revenue,
            'profit': profit
        }

# Create the summary DataFrame
summary_df = pd.DataFrame(client_totals).T
summary_df.index.name = 'Client ID'
summary_df.sort_values('profit', ascending=False, inplace=True)

# Display the summary DataFrame
print(summary_df)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

summary_df['total_units'] = summary_df['total_units'].astype(int)
summary_df['shipping_price'] = (summary_df['shipping_price'] / 1000000).round(2)
summary_df['revenue'] = (summary_df['revenue'] / 1000000).round(2)
summary_df['profit'] = (summary_df['profit'] / 1000000).round(2)

# Rename the columns
summary_df = summary_df.rename(columns={
    'total_units': 'Total Units Purchased (in units)',
    'shipping_price': 'Total Shipping Price (in millions of dollars)',
    'revenue': 'Total Revenue (in millions of dollars)',
    'profit': 'Total Profit (in millions of dollars)'
})

# Display the formatted and renamed summary DataFrame
print(summary_df)
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
summary_df = summary_df.sort_values('Total Profit (in millions of dollars)', ascending=False)

# Display the sorted summary DataFrame
print(summary_df)
```
